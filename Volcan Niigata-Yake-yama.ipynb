```python
position_volcan = {'utm_est': [4089872],
        'utm_nord': [236226]}

# Create DataFrame
df_position_volcan = pd.DataFrame(position_volcan)
df_position_volcan
```

```python

stations['r']= np.sqrt((np.square(df_position_volcan['utm_nord'][0]-stations['Eastern']))+(np.square(df_position_volcan['utm_est'][0]-stations['Northern'])))
```

```python
stations['Ur']=np.sqrt((np.square(stations['U_east']))+np.square(stations['U_north']))
stations

#ici la valeur de Uz est quivalent  la valeur de U_up 
#h correspond  l'elevation 
```

```python
#Calcul de f, la profondeur sous le niveau de la mer (en m)
stations['f']=((stations['r']*stations['U_up'])/stations['Ur'])-stations['Elevation']
stations
```

```python
#Calcul de la variation de volume dans le reservoir appele V (en m^3)

stations['V']=(4*np.pi*(np.power(np.square(stations['r']*stations['U_up']/stations['Ur'])+np.square(stations['r']),(3/2)))*stations['Ur'])/(3*stations['r'])
```

```python
#On cre une fonction qui nous donne les carts les plus proches de zro (une valeur ngative et une valeur positive)
#Ces deux valeurs seront les brones de notre prochain intervale pour "zoomer" et obtenir des valeurs plus prcises de V et f




def ecart_proche_de_zero (liste_ecart,liste_V,liste_f):
    ecarts_droit = 0000.1
    ecarts_gauche = -0000.1
    valeur_V=None
    valeur_f=None
    valeur_1_V = None
    valeur_1_f = None
    n=0
    
    for i in liste_ecart :
        
        n=n+1
        
        if i >= 0 :
            if i <= ecarts_droit:
                ecarts_droit=i
                valeur_V = liste_V[n]
                valeur_f = liste_f[n]
                
        if i <= 0 :
            if i >= ecarts_gauche:
                ecarts_gauche=i
                valeur_1_V = liste_V[n]
                valeur_1_f = liste_f[n]
    return ecarts_gauche, ecarts_droit, valeur_V,valeur_f,valeur_1_V,valeur_1_f



#Cette fonction a pour but de trouver les deux valeurs les plus proches de zro dans une liste (positive et ngative)."""
#Puisqu'il y aura deux valeurs, une positive et une ngative, notre fonction aura deux parties.

    
```

```python

result1=[]
result2=[]
# SUPPRIMER DECLARATION 
ecarts=[]
storage_variable1=[]

valeur = 0.0001
valeurneg = -0.0001
 

for V in np.linspace(0,60e6,2000):
    for f in np.linspace(-10e3,0,200):
        Uz=((3*V*(f+stations['Elevation'][0]))/(4*np.pi*(np.power((np.square(f+stations['Elevation'][0])+np.square(stations['r'][0])),(3/2)))))

        variable = stations['U_up'][0] - Uz

        #expliquer float 

        if valeurneg <= variable <= valeur  :
            ecarts=variable
            storage_variable1.append(float(ecarts))
            result1.append(float(V))
            result2.append(float(f))
            

print(ecart_proche_de_zero(storage_variable1,result1,result2))


plt.plot(result1,result2)
plt.show()

#itration 2

result3_1=[]
result4_1=[]
storage_variable=[]



valeur = ecart_proche_de_zero(storage_variable1,result1,result2)[1]
valeurneg= ecart_proche_de_zero(storage_variable1,result1,result2)[0]



for V in np.linspace(ecart_proche_de_zero(storage_variable1,result1,result2)[2],ecart_proche_de_zero(storage_variable1,result1,result2)[4],2000):
    for f in np.linspace(ecart_proche_de_zero(storage_variable1,result1,result2)[3],ecart_proche_de_zero(storage_variable1,result1,result2)[5],200):

    
        Uz=((3*V*(f+stations['Elevation'][0]))/(4*np.pi*(np.power((np.square(f+stations['Elevation'][0])+np.square(stations['r'][0])),(3/2)))))

        variable = stations['U_up'][0] - Uz


        #expliquer float 

        if valeurneg <= variable <= valeur  :
            ecarts=variable
            storage_variable.append(float(ecarts))
            result3_1.append(float(V))
            result4_1.append(float(f))



print (ecart_proche_de_zero(storage_variable,result3_1,result4_1))

plt.plot(result3_1,result4_1)
plt.show()

#arreter l'itaration parcque la prochaine de permet d'obetenir que des nan
    
```
